```python
import numpy as np
import matplotlib.pyplot as plt


data = np.random.randint(1, 100, 50)

mean = np.mean(data)
min_value = np.min(data)
max_value = np.max(data)

x_values = np.arange(1, len(data)+1)

plt.figure(figsize=(8, 6))

plt.plot(x_values, data, label='Data', marker='o', linestyle='--', color='blue')
plt.axhline(mean, color='green', linestyle='-', label=f'Mean: {mean:.2f}')
plt.axhline(min_value, color='red', linestyle='--', label=f'Min: {min_value}')
plt.axhline(max_value, color='purple', linestyle='--', label=f'Max: {max_value}')

plt.title('Line Plot of Data with Mean, Min, and Max')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()

plt.show()
```
